# Pattern 4 : Two Pointers

**Problem:** Remove duplicates from sorted array or find pair with given sum

**Solution Approach:**
Use two pointers that move towards each other or in the same direction based on the problem. Common patterns: opposite ends (for sorted array), same direction (for removing duplicates).

## Python Code:

```python
def two_sum_sorted(arr, target):
    """
    Find pair in sorted array that sums to target

    Args:
        arr: Sorted list of integers
        target: Target sum
    Returns:
        Indices of two numbers, or [-1, -1] if not found
    Example:
        >>> two_sum_sorted([1, 2, 3, 4, 6], 6)
        [1, 3] # arr[1] + arr[3] = 2 + 4 = 6
    """

    left, right = 0, len(arr) - 1

    while left < right:
        curr_sum = arr[left] + arr[right]

        if curr_sum == target:
            return [left, right]
        elif curr_sum < target:
            left += 1
        else:
            right -= 1
    
    return [-1, -1]
```

```python
def remove_duplicates(arr):
    if not arr:
        return 0
    
    write_idx = 1

    for read_idx in range(1, len(arr)):
        if arr[read_idx] != arr[read_idx - 1]:
            arr[write_idx] = arr[read_idx]
            write_idx += 1
    
    return write_idx
```

```python
# My leetcode Problem : 2
def removeDuplicates(nums) -> int:
    i = 0
    n = len(nums)
    for j in range(1, n):
        if nums[i] != nums[j]:
            i += 1
            nums[i] = nums[j]

    return i + 1 
```

```python
def three_sum(nums):
    """
    Find all unique triplets that sum to zero.
    Combines sorting + two_pointer technique

    Returns:
        List of triplets [a, b, c] such that a + b + c = 0
    """
    nums.sort()
    res = []

    for i in range(len(nums) - 2):
        # Skip duplicate for first number
        if i > 0 and nums[i] == nums[i-1]:
            continue

        # Two pointers for remaining array
        left, right = i + 1, len(nums) - 1
        target = -nums[i]

        while left < right:
            curr_sum = nums[left] + nums[right]

            if curr_sum == target:
                res.append([nums[i], nums[left], nums[right]])

                while left < right and nums[left] == nums[left + 1]:
                    left += 1
                while left < right and nums[right] == nums[right - 1]:
                    right -= 1
                
                left += 1
                right -= 1
            
            elif curr_sum < target:
                left += 1
            else:
                right -= 1
    
    return res
```

```python
def is_palindrome(s):
    l, r = 0, len(s) - 1

    while l < r:
        while l < r and not s[l].isalnum():
            l += 1
        while l < r and not s[r].isalnum():
            r -= 1
        
        if s[l].lower() != s[r].lower():
            return False
        
        l += 1
        r -= 1

    return True
```

```python
# Example Usage
print("Two Sum in Sorted Array:")
arr = [1, 2, 3, 4, 6]
target = 6
result = two_sum_sorted(arr, target)
print(f"Array: {arr}, Target: {target}")
print(f"Indices: {result}")
```

```python
print("Remove Duplicates:")
nums = [1, 1, 2, 2, 3, 4, 4]
print(f"Original: {nums}")
length = remove_duplicates(nums)
print(f"After removing duplicates: {nums[:length]}")
```

```python
print("Three Sum: ")
nums = [-1, 0, 1, 2, -1, 4]
result = three_sum(nums)
print(f"Array: {nums}")
print(f"Triplets that sum to 0: {result}")
```

```python
print("Palindrome Check:")
test_str = "A man, a plan, a canal: Panama"
print(f"String: {test_str}")
print(f"Is palindrome: {is_palindrome(test_str)}")
```

**Complexity Analysis** : 
* **Time Complexity:** ${O(n)}$ for two pointers, ${O(n^2)}$ for three sum
* **Space Complexity:** ${O(1)}$ excluding output
