# Top K elements (Heap)

**Problem:** Find K largest/smallest elements

## Python Code:

```python
import heapq

def find_k_largest(nums, k):
    """
    Find K largest elements using min heap.
    Example:
    >>> find_k_largest([3, 2, 1, 5, 6, 4], 2)
    [5, 6]
    """
    # Use min_heap of size k
    heap = []

    for num in nums:
        heapq.heappush(heap, num)
        if len(heap) > k:
            heapq.heappop(heap) # Remove smallest
    
    return heap
```

```python
def kth_largest_element(nums, k):
    """
    Find Kth largest element.
    Example:
    >>> kth_largest_element([3,2,1,5,6,4], 2)
    5 # 2nd largest
    """
    heap = []

    for num in nums:
        heapq.heappush(heap, num)
        if len(heap) > k:
            heapq.heappop(heap)
    
    return heap[0] # root for min heap
```

```python
from collections import Counter
def top_k_frequent(nums, k):
    count = Counter(nums)

    return heapq.nlargest(k, count.keys(), key=count.get)
```

```python
# Examples
nums = [3, 2, 1, 5, 6, 4]
k = 2
print(f"Array: {nums}")
print(f"Top {k} largest: {find_k_largest(nums, k)}")
print(f"{k}th largest element: {kth_largest_element(nums, k)}")
nums2 = [1,1,1,2,2,3]
k2 = 2
print(f"\nArray: {nums2}")
print(f"Top {k2} frequent: {top_k_frequent(nums2, k2)}")
```

### Complexity : ${O(n*log(k))}$ - Heap Operations
