# Pattern 2: Sliding Window

**Problem:** Find the maximum sum of k consecutive elements in an array.

**Solution Approach:**
Instead of recalculating the sum for each window from scratch, maintain a window and slide it by removing the leftmost element and adding the next element. This reduces time from ${O(n*k)}$ to ${O(n)}$.

## Python Code:

```python
def max_sum_subarray(arr, k):
    """
    Find maximum sum of k consecutive elements.
    Args:
    arr: List of integers
    k: Window size
    Returns:
    Maximum sum of k consecutive elements, or None if invalid
    Example:
    >>> max_sum_subarray([1, 4, 2, 10, 23, 3, 1, 0, 20], 4)
    39 # [4, 2, 10, 23]
    """
    # Validation
    if not arr or len(arr) < k or k <= 0:
        return None
    # Calculate sum of first window
    window_sum = sum(arr[:k])
    max_sum = window_sum
    # Slide the window
    for i in range(k, len(arr)):
        # Remove leftmost element, add rightmost element
        window_sum = window_sum - arr[i-k] + arr[i]
        max_sum = max(max_sum, window_sum)
    return max_sum
```

```python
# Advanced: Variable-size window
def longest_subarray_with_sum_k(arr, k):
    """
    Find length of longest subarray with sum <= k.
    This demonstrates a variable-size sliding window.
    """
    left = 0
    current_sum = 0
    max_length = 0

    for right in range(len(arr)):
        current_sum += arr[right]

        # Shrink window if sum exceeds k
        while current_sum > k and left <= right:
            current_sum -= arr[left]
            left += 1
        
        max_length = max(max_length, right - left + 1)

    return max_length
```

```python
# Example usage

print("Fixed Window Example:")
arr = [1, 4, 2, 10, 23, 3, 1, 0, 20]
k = 4
result = max_sum_subarray(arr, k)
print(f"Array: {arr}")
print(f"k = {k}")
print(f"Maximum sum of {k} consecutive elements: {result}")
print("\nVariable Window Example:")
arr2 = [1, 2, 3, 4, 5]
k2 = 11
result2 = longest_subarray_with_sum_k(arr2, k2)
print(f"Array: {arr2}")
print(f"Sum <= {k2}")
print(f"Longest subarray length: {result2}")
```

**Complexity Analysis:**
* **Time Complexity:** ${O(n)}$ - Single pass through array
* **Space Complexity:** ${O(1)}$ - Only storing window sum and max

**When to Use:**
* Finding optimal subarray of fixed size
* Maximum/minimum of contiguous elements
* Longest substring with certain property
