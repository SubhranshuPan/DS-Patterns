# Pattern 6: Recursion with Memoization (Dynamic Programming)

**Problem:** Calculate Fibonacci numbers efficiently.

**Solution Approach:**
Store results of expensive function calls and return cached result when same inputs occur again. This prevents redundant calculations in recursive solutions.

## Python Code:

```python
def fibonacci_naive(n):
    """
    Very slow for large n
    Time: O(2^n) - exponential !
    """
    if n <= 1:
        return n    
    return fibonacci_naive(n - 1) + fibonacci_naive(n - 2)
```

```python
def fibonacci_memo(n, memo=None):
    """
    Fibonacci with memoization
    Time : O(n), Space : O(n)
    Example:
        >>> fibonacci_memo(10)
        55
    """
    if memo is None:
        memo = {}

    if n <= 1:
        return n

    if n in memo:
        return memo[n]
    
    memo[n] = fibonacci_memo(n-1, memo) + fibonacci_memo(n-2, memo)
    return memo[n]
```

```python
# Using @lru_cache_decorator 
from functools import lru_cache

@lru_cache(maxsize=None)
def fibonacci_cached(n):
    """
    Fibonacci using Python's built-in caching 
    """
    if n <= 1:
        return n
    return fibonacci_cached(n-1) + fibonacci_cached(n-2)
```

```python
# Examples
import time
print("Fibonacci Comparison:")
n = 35
start = time.time()
result_naive = fibonacci_naive(n) # Too slow, commented out
print(f"Naive: {result_naive}, Time: {time.time() - start:.4f}s")
start = time.time()
result_memo = fibonacci_memo(n)
print(f"Memoized: {result_memo}, Time: {time.time() - start:.6f}s")
```

```python
def climbing_stairs(n, memo=None):
    """
    You can climb 1 or 2 steps. How many ways to reach top?
    This is actually Fibonacci in disguise!
    Args:
    n: Number of steps
    Returns:
    Number of distinct ways to climb to top
    Example:
    >>> climbing_stairs(3)
    3 # [1,1,1], [1,2], [2,1]
    """
    if memo is None:
        memo = {}
    
    if n <= 2:
        return n
    
    if n in memo:
        return memo[n]
    
    memo[n] = climbing_stairs(n-1, memo) + climbing_stairs(n-2, memo)
    return memo[n]
```

```python
print("Climbing Stairs:")
n = 5
result = climbing_stairs(n)
print(f"Steps: {n}, Ways to climb: {result}")
```

```python
def min_cost_climbing_stairs(cost, memo=None):
    """
    Minimum cost to reach top when each step has a cost.
    Args:
    cost: List where cost[i] is cost of step i
    Returns:
    Minimum cost to reach top
    """
    if memo is None:
        memo = {}

    def dp(i):
        if i <= 1:
            return cost[i]
        
        if i in memo:
            return memo[i]

        memo[i] = cost[i] + min(dp(i-1), dp(i-2))
        return memo[i]

    n = len(cost)
    return min(dp(n-1), dp(n-2)) 
```

```python
print("Min Cost Climbing Stairs:")
cost = [10, 15, 20]
result = min_cost_climbing_stairs(cost)
print(f"Costs: {cost}, Minimum cost: {result}")
```

```python
def lcs(t1, t2, i=None, j=None, memo=None):
    if i is None:
        i = 0
    if j is None:
        j = 0
    if memo is None:
        memo = {}
    
    if i == len(t1) or j == len(t2):
        return 0

    if (i, j) in memo:
        return memo[(i, j)]
    
    if t1[i] == t2[j]:
        memo[(i, j)] = 1 + lcs(t1, t2, i+1, j+1, memo)
    else:
        memo[(i, j)] = max(
            lcs(t1, t2, i+1, j, memo),
            lcs(t1, t2, i, j+1, memo)
        )

    return memo[(i, j)]
```

```python
print("Longest Common Subsequence:")
text1, text2 = "abcde", "ace"
result = lcs(text1, text2)
print(f"Text1: {text1}, Text2: {text2}")
print(f"LCS length: {result}")
```

## Complexity Analysis:
* **Without Memoization:** ${O(n^2)}$ - exponential time
* **With Memoization:** ${O(n)}$ - Time, ${O(n)}$ - Space

## When to Use:
* Overlapping Subproblems (same calculation repeated)
* Optimal substructure (soltion built from subproblems)
* Fibonacci, Climbing stairs, LCS, Knapsack problems
