### Pattern 10 : BFS Tree Traversal

```python
from collections import deque

def level_order_traversal(root):
    if not root:
        return []
    
    result = []
    queue = deque([root])

    while queue:
        level_size = len(queue)
        curr_level = []

        for _ in range(level_size):
            node = queue.popleft()
            curr_level.append(node.val)

            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)

        result.append(curr_level)

    return result
```

```python
# --- Example Usage ---

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

# Build a sample binary tree:
#         1
#        / \
#       2    3
#      / \    \
#     4   5    6

root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
root.right.right = TreeNode(6)

result = level_order_traversal(root)

print('Level Order Traversal:')
for i, level in enumerate(result):
    print(f'  Level {i}: {level}')

# Output:
#   Level 0: [1]
#   Level 1: [2, 3]
#   Level 2: [4, 5, 6]
```

### Pattern 11 : DFS Graph Search

```python
def dfs_recursive(graph, node, visited=None):
    """
    Depth-First Search using recursion.
    
    Args:
        graph: adjacency list (dict of node -> list of neighbors)
        node: starting node
        visited: set of already-visited nodes
    
    Returns:
        list of nodes in DFS visit order
    """
    if visited is None:
        visited = set()
    
    visited.add(node)
    print(node, end=' ')

    for neighbor in graph[node]:
        if neighbor not in visited:
            result.extend(dfs_recursive(graph, neighbor, visited))
    
    return result
```

```python
def dfs_iterative(graph, start):
    """
    Depth-First Search using an explicit stack.
    
    Args:
        graph: adjacency list (dict of node -> list of neighbors)
        start: starting node
    
    Returns:
        list of nodes in DFS visit order
    """

    visited = set()
    stack = [start]
    result = []

    while stack:
        node = stack.pop()
        if node not in visited:
            visited.add(node)
            result.append(node)
            print(node, end=' ')
            # Push neighbors in reverse order so that
            # the left-most neighbor is processed first
            for neighbor in reversed(graph[node]):
                if neighbor not in visited:
                    stack.append(neighbor)
    return result
```

```python
# Sample graph (adjacency list):
#   0 --- 1 --- 3
#   |     |
#   2     4

graph = {
    0: [1, 2],
    1: [0, 3, 4],
    2: [0],
    3: [1],
    4: [1]
}

print("Recursive DFS:", dfs_recursive(graph, 0))
print("Iterative DFS:", dfs_iterative(graph, 0))

# Output:
#   Recursive DFS: [0, 1, 3, 4, 2]
#   Iterative DFS: [0, 1, 3, 4, 2]
```
