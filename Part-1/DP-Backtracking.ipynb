## Pattern 12: Dynamic Programming (Bottom up)

```python
def coin_change(coins, amount):
    """Minimum coins to make amount"""
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0

    for coin in coins:
        for i in range(coin, amount + 1):
            dp[i] = min(dp[i], dp[i-coin] + 1)
    
    return dp[amount] if dp[amount] != float('inf') else -1
```

### Time Complexity : ${O(C * A)}$
${C}$ = number of coin denominations
${A}$ = Amount
Two nested loops: outer over coins, inner over amounts.

```python
def knapsack(weights, values, capacity):
    """0/1 Knapsack problem"""
    n = len(weights)
    dp = [[0] * (capacity + 1) for _ in range(n+1)]

    for i in range(1, n + 1):
        for w in range(capacity+1):
            if weights[i-1] <= w:
                dp[i][w] = max(
                    values[i-1] + dp[i-1][w - weights[i-1]],
                    dp[i-1][w]
                )
            else:
                dp[i][w] = dp[i-1][w]

    return dp[n][capacity]
```

```python
coins = [1, 5, 10, 25]
amount = 30

print(coin_change(coins, amount))  # Output: 2
```

```python
weights  = [2, 3, 4, 5]
values   = [3, 4, 5, 6]
capacity = 8

print(knapsack(weights, values, capacity))  # Output: 10
```

# Pattern 13 : Backtracking

```python
def subsets(nums):
    result = []

    def backtrack(start, path):
        result.append(path[:])
        for i in range(start, len(nums)):
            path.append(nums[i])
            backtrack(i + 1, path)
            path.pop()
    
    backtrack(0, [])
    return result
```

```python
nums = [1, 2, 3]
print(subsets(nums))
```
