## Pattern 14: Greedy Algorithm

### 🧠 What is a Greedy Algorithm?

A **Greedy Algorithm** builds a solution **step by step**, always choosing the option that looks **best right now** (the locally optimal choice), hoping it leads to the **globally optimal** solution.

> **Think of it like eating at a buffet** 🍽️ — at each station, you pick the dish that looks most appealing *right now*, without worrying about what's coming next.

---

### 🔑 Key Characteristics

| Property | Description |
|---|---|
| **Greedy Choice** | At each step, pick the *locally optimal* option |
| **Optimal Substructure** | An optimal solution contains optimal solutions to sub-problems |
| **No Backtracking** | Once a choice is made, it is **never reconsidered** |
| **Not Always Correct** | Greedy works only when the problem has the *greedy-choice property* |

---

### 📐 When Does Greedy Work?

A problem is suitable for a greedy approach when:
1. **Greedy-Choice Property** — A locally optimal choice leads to a globally optimal solution.
2. **Optimal Substructure** — The problem can be broken into sub-problems, and solving each optimally gives the overall optimal answer.

> ⚠️ If a problem lacks these properties, greedy may give a **wrong answer** — use Dynamic Programming instead.

---

### 🔄 Generic Greedy Algorithm Steps
1. SORT or ORGANIZE the input based on a greedy criterion
2. INITIALIZE the solution (empty set, counter, etc.)
3. FOR each element in the sorted input: 
    * CHECK if this element is compatible / feasible 
    * If YES → include it in the solution 
    * If NO → skip it
4. RETURN the final solution

### 🪜 Step-by-Step Walkthrough: Activity Selection Problem

**Problem:** Given `n` activities with start and finish times, select the **maximum number of non-overlapping activities**.

**Greedy Strategy:** Always pick the activity that **finishes earliest** — this leaves the most room for future activities.

---

#### Example Input

| Activity | Start | Finish |
|----------|-------|--------|
| A        | 1     | 3      |
| B        | 2     | 5      |
| C        | 4     | 7      |
| D        | 1     | 8      |
| E        | 5     | 9      |
| F        | 8     | 10     |

---

#### Step-by-Step Execution

**Step 1 — Sort by finish time:**

| Activity | Start | Finish |
|----------|-------|--------|
| A        | 1     | 3  ✅  |
| B        | 2     | 5      |
| C        | 4     | 7      |
| D        | 1     | 8      |
| E        | 5     | 9      |
| F        | 8     | 10     |

**Step 2 — Select A** (first activity, always selected). Last finish = `3`

**Step 3 — Check B** (start=2): `2 < 3` → ❌ overlaps → skip

**Step 4 — Check C** (start=4): `4 >= 3` → ✅ select! Last finish = `7`

**Step 5 — Check D** (start=1): `1 < 7` → ❌ overlaps → skip

**Step 6 — Check E** (start=5): `5 < 7` → ❌ overlaps → skip

**Step 7 — Check F** (start=8): `8 >= 7` → ✅ select! Last finish = `10`

---

**Result:** Selected **3 activities** → `A(1,3)`, `C(4,7)`, `F(8,10)`



### 🐍 Code Walkthrough

The function below implements the Activity Selection problem using the greedy approach:


```python
def activity_selection(start, finish):
    """Select maximum activities"""
    activities = sorted(zip(start, finish), key=lambda x: x[1])
    selected = [activities[0]]

    for activity in activities[1:]:
        if activity[0] >= selected[-1][1]:
            selected.append(activity)
    
    return len(selected)
```

### ▶️ Example Usage


```python
# Example: Activity Selection
start_times  = [1, 2, 4, 1, 5, 8]
finish_times = [3, 5, 7, 8, 9, 10]

result = activity_selection(start_times, finish_times)
print(f"Maximum non-overlapping activities: {result}")

# Detailed output version
def activity_selection_detailed(start, finish):
    """Select maximum activities with detailed output"""
    activities = sorted(zip(start, finish), key=lambda x: x[1])
    selected = [activities[0]]
    
    print(f"Sorted activities (by finish time): {activities}\n")
    print(f"Step 1: Select activity {activities[0]}")
    
    for i, activity in enumerate(activities[1:], 2):
        if activity[0] >= selected[-1][1]:
            selected.append(activity)
            print(f"Step {i}: Activity {activity} starts at {activity[0]} >= {selected[-2][1]} → ✅ Selected")
        else:
            print(f"Step {i}: Activity {activity} starts at {activity[0]} <  {selected[-1][1]} → ❌ Skipped")
    
    print(f"\nSelected activities: {selected}")
    return len(selected)

print("\n--- Detailed Trace ---\n")
count = activity_selection_detailed(start_times, finish_times)
print(f"\nTotal selected: {count}")
```

### ⏱️ Time & Space Complexity

#### Time Complexity: `O(n log n)`

| Step | Operation | Cost |
|------|-----------|------|
| 1    | Sorting activities by finish time | `O(n log n)` |
| 2    | Single pass through sorted list   | `O(n)` |
| **Total** | | **`O(n log n)`** |

> The **sorting step dominates**. If the input is already sorted, it reduces to `O(n)`.

#### Space Complexity: `O(n)`

| Component | Space |
|-----------|-------|
| Sorted copy of activities | `O(n)` |
| Selected activities list  | `O(n)` worst case |
| **Total** | **`O(n)`** |

> Can be optimized to `O(1)` extra space if we sort in-place and only track the count.

---

### 🔁 Greedy vs Dynamic Programming

| Aspect | Greedy | Dynamic Programming |
|--------|--------|---------------------|
| **Approach** | Makes one choice at each step | Considers all choices |
| **Speed** | Generally faster | Slower but more thorough |
| **Correctness** | Only for problems with greedy-choice property | Works for all optimization problems with optimal substructure |
| **Backtracking** | Never | Revisits sub-problems |
| **Example** | Activity Selection, Huffman Coding | 0/1 Knapsack, Longest Common Subsequence |

---

### 🌍 Other Classic Greedy Problems

1. **Fractional Knapsack** — Take fractions of items by value/weight ratio
2. **Huffman Encoding** — Build optimal prefix-free codes
3. **Dijkstra's Algorithm** — Shortest path in weighted graphs
4. **Prim's / Kruskal's** — Minimum Spanning Tree
5. **Job Scheduling** — Maximize profit with deadlines

---

### 💡 Key Takeaway

> **Greedy = "Make the best local choice and never look back."**
> It's fast and elegant — but only correct when the problem guarantees that local optimality leads to global optimality.
