## Pattern 17: Bit Manipulation

### 🧠 What is Bit Manipulation?

**Bit Manipulation** is the technique of operating directly on the **binary representation** of numbers using **bitwise operators**. Instead of doing arithmetic on decimal values, you work with the individual **0s and 1s** that make up a number in memory.

> **Think of it like flipping light switches** 💡 — Each bit is a tiny switch (ON=1, OFF=0). Bit manipulation lets you flip, check, or combine switches incredibly fast — in a single CPU instruction!

---

### 🔑 The Bitwise Operators

| Operator | Symbol | Description | Example (`5=101`, `3=011`) |
|----------|--------|-------------|---------------------------|
| **AND** | `&` | Both bits must be 1 | `101 & 011 = 001` → 1 |
| **OR** | `\|` | At least one bit is 1 | `101 \| 011 = 111` → 7 |
| **XOR** | `^` | Bits must be different | `101 ^ 011 = 110` → 6 |
| **NOT** | `~` | Flip all bits | `~101 = ...010` |
| **Left Shift** | `<<` | Shift bits left (×2) | `101 << 1 = 1010` → 10 |
| **Right Shift** | `>>` | Shift bits right (÷2) | `101 >> 1 = 10` → 2 |

---

### ⚡ Essential Bit Tricks

These are the building blocks used in almost every bit manipulation problem:

```
1. Check if last bit is 1:     n & 1          → gives 1 if odd, 0 if even
2. Turn off last set bit:      n & (n - 1)    → removes rightmost 1
3. Isolate last set bit:       n & (-n)       → keeps only rightmost 1
4. Check power of 2:           n & (n - 1) == 0  (n > 0)
5. XOR with itself:            a ^ a = 0      → any number XOR itself is 0
6. XOR with zero:              a ^ 0 = a      → any number XOR zero is itself
7. Swap without temp:          a ^= b; b ^= a; a ^= b
```

---

### 🔑 Why `n & (n - 1)` Removes the Last Set Bit

This is the most important trick to understand:

```
Example: n = 12 (binary: 1100)

  n     = 1100
  n - 1 = 1011    ← subtracting 1 flips all bits from the rightmost 1 onward
  ─────────────
  n & (n-1) = 1000   ← the rightmost 1 is gone!

If n is a power of 2 (only ONE bit set):
  n     = 1000   (8)
  n - 1 = 0111   (7)
  ─────────────
  n & (n-1) = 0000   ← result is 0! That's how we detect power of 2.
```

---

### 🔑 Why XOR Finds the Unique Number

XOR has two magical properties:

```
Property 1: a ^ a = 0     (paired numbers cancel out)
Property 2: a ^ 0 = a     (zero is identity)

Example: nums = [4, 1, 2, 1, 2]

  4 ^ 1 ^ 2 ^ 1 ^ 2
= 4 ^ (1 ^ 1) ^ (2 ^ 2)     ← rearrange (XOR is commutative)
= 4 ^ 0 ^ 0                   ← pairs cancel to 0
= 4                            ← only the unique number survives!
```

#### Function 1: `count_set_bits(13)` — Count the 1s in binary

`13` in binary = `1101` → expected answer: **3** set bits

```
Iteration 1:
  n = 1101 (13)
  n & 1 = 1101 & 0001 = 1   → count = 1   (last bit is 1)
  n >>= 1 → n = 0110 (6)

Iteration 2:
  n = 0110 (6)
  n & 1 = 0110 & 0001 = 0   → count = 1   (last bit is 0, skip)
  n >>= 1 → n = 0011 (3)

Iteration 3:
  n = 0011 (3)
  n & 1 = 0011 & 0001 = 1   → count = 2   (last bit is 1)
  n >>= 1 → n = 0001 (1)

Iteration 4:
  n = 0001 (1)
  n & 1 = 0001 & 0001 = 1   → count = 3   (last bit is 1)
  n >>= 1 → n = 0000 (0)

n == 0 → STOP. Return count = 3 ✅
```

---

#### Function 2: `is_power_of_two(n)`

```
n = 8  → binary: 1000
  n > 0? ✅
  n & (n-1) = 1000 & 0111 = 0000 == 0? ✅ → True (8 is 2³)

n = 6  → binary: 0110
  n > 0? ✅
  n & (n-1) = 0110 & 0101 = 0100 == 0? ❌ → False (6 has multiple set bits)

n = 1  → binary: 0001
  n > 0? ✅
  n & (n-1) = 0001 & 0000 = 0000 == 0? ✅ → True (1 is 2⁰)

n = 0  → n > 0? ❌ → False (0 is NOT a power of 2)
```

---

#### Function 3: `single_number([2, 3, 2, 4, 3])`

```
Start: result = 0           binary: 00000

Step 1: result ^= 2
  00000 ^ 00010 = 00010     result = 2

Step 2: result ^= 3
  00010 ^ 00011 = 00001     result = 1

Step 3: result ^= 2
  00001 ^ 00010 = 00011     result = 3

Step 4: result ^= 4
  00011 ^ 00100 = 00111     result = 7

Step 5: result ^= 3
  00111 ^ 00011 = 00100     result = 4 ✅

The pairs (2,2) and (3,3) cancel out → only 4 remains!
```

### 🐍 Code Walkthrough

#### `count_set_bits(n)` — Brian Kernighan's approach (simplified)
- Uses `n & 1` to check if the **last bit** is 1 (adds to count if so)
- Right-shifts `n` by 1 (`n >>= 1`) to move to the next bit
- Repeats until `n` becomes 0

#### `is_power_of_two(n)`
- A power of 2 has **exactly one** set bit: `1, 10, 100, 1000...`
- `n & (n - 1)` removes the last set bit — if the result is 0, there was only one bit → power of 2
- Also checks `n > 0` since 0 is not a power of 2

#### `single_number(nums)` — XOR trick
- XOR all numbers together
- Duplicate numbers cancel out (`a ^ a = 0`)
- The unique number is left standing (`a ^ 0 = a`)

````python
def count_set_bits(n):
    """Count 1s in Binary"""
    count = 0
    while n:
        count += n & 1
        n >>= 1
    return count
````

````python
def is_power_of_two(n):
    """Check if n is power of 2"""
    return n > 0 and (n & (n - 1)) == 0
````

````python
def single_number(nums):
    """Find number appearing once. (XOR)"""
    result = 0
    for num in nums:
        result ^= num
    return result
````

### ▶️ Example Usage

````python
# === count_set_bits ===
print("=== count_set_bits ===\n")
test_nums = [0, 1, 7, 13, 16, 255]
for n in test_nums:
    bits = count_set_bits(n)
    print(f"  count_set_bits({n:>3}) = {bits}    binary: {bin(n)}")

# === is_power_of_two ===
print("\n=== is_power_of_two ===\n")
test_pows = [0, 1, 2, 3, 4, 8, 15, 16, 64, 100]
for n in test_pows:
    result = is_power_of_two(n)
    symbol = "✅ Yes" if result else "❌ No"
    print(f"  is_power_of_two({n:>3}) → {symbol}    binary: {bin(n)}")

# === single_number ===
print("\n=== single_number ===\n")
test_arrays = [
    [2, 3, 2, 4, 3],
    [1],
    [4, 1, 2, 1, 2],
    [7, 3, 5, 3, 7],
]
for arr in test_arrays:
    result = single_number(arr)
    print(f"  single_number({arr}) → {result}")

# === Bonus: Visualize bit operations ===
print("\n=== Visualizing Bit Operations ===\n")
a, b = 13, 10
print(f"  a = {a:>3}  →  {a:08b}")
print(f"  b = {b:>3}  →  {b:08b}")
print(f"  {'─' * 30}")
print(f"  a & b  = {a&b:>3}  →  {a&b:08b}   (AND)")
print(f"  a | b  = {a|b:>3}  →  {a|b:08b}   (OR)")
print(f"  a ^ b  = {a^b:>3}  →  {a^b:08b}   (XOR)")
print(f"  a << 1 = {a<<1:>3}  →  {a<<1:08b}   (Left Shift)")
print(f"  a >> 1 = {a>>1:>3}  →  {a>>1:08b}   (Right Shift)")
````

### ⏱️ Time & Space Complexity

#### `count_set_bits(n)`

| Aspect | Complexity | Explanation |
|--------|-----------|-------------|
| **Time** | **`O(log n)`** or **`O(k)`** | Loops once per bit in `n`. A number `n` has `⌊log₂n⌋ + 1` bits. Also expressed as `O(k)` where k = number of bits |
| **Space** | **`O(1)`** | Only uses a counter variable |

----
> 💡 **Brian Kernighan's optimization:** Replace the loop body with `n &= (n - 1)` and it runs in `O(set bits)` instead of `O(total bits)` — skips the 0 bits entirely!

#### `is_power_of_two(n)`

| Aspect | Complexity | Explanation |
|--------|-----------|-------------|
| **Time** | **`O(1)`** | Single comparison + single bitwise AND — constant time! |
| **Space** | **`O(1)`** | No extra space |

#### `single_number(nums)`

| Aspect | Complexity | Explanation |
|--------|-----------|-------------|
| **Time** | **`O(n)`** | Single pass through the array, XOR each element once |
| **Space** | **`O(1)`** | Only one variable `result` — no hash maps or sets needed! |

----
> This is what makes the XOR solution elegant — hash map approach also takes `O(n)` time but needs `O(n)` space. XOR does it in `O(1)` space!

### 🌍 Classic Bit Manipulation Problems

| Problem | Technique |
|---------|-----------|
| **Number of 1 Bits** | `n & 1` + right shift (or `n & (n-1)`) |
| **Power of Two** | `n & (n - 1) == 0` |
| **Single Number** | XOR all elements |
| **Single Number II** (appears 3×) | Bit counting modulo 3 |
| **Missing Number** | XOR indices with values |
| **Reverse Bits** | Shift and build result bit by bit |
| **Counting Bits (0 to n)** | DP + `i & (i-1)` |
| **Subsets** | Use bitmask `0` to `2ⁿ-1` to represent subsets |

---

### 💡 Key Takeaway

> **Bit Manipulation = "Solve problems at the binary level for O(1) space and blazing speed."**
> The three essential tricks are: `n & 1` (check last bit), `n & (n-1)` (remove last set bit), and `a ^ a = 0` (cancel duplicates). Master these and most bit problems become intuitive!
