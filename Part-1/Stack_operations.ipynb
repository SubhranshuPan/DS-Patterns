# Valid Parenthesis

```python
def isValid_parenthesis(s):
    stack = []
    mapping = {')' : '(', '}' : '{', ']' : '['}

    for char in s:
        if char in mapping:
            if not stack or stack[-1] != mapping[char]:
                return False
            stack.pop()
        else:
            stack.append(char)
    return len(stack) == 0
```

```python
isValid_parenthesis("(){}[]")
```

```python
isValid_parenthesis("({[)]}")
```
