# Pattern 7 : Merge Intervals

**Problem:** Overlapping Intervals

## Python Code:

```python
def merge_intervals(intervals):
    """
    Merge overlapping intervals.
    Example:
    >>> merge_intervals([[1,3], [2,6], [8,10], [15,18]])
    [[1,6], [8,10], [15,18]]
    """
    if not intervals:
        return []
    
    intervals.sort(key=lambda x: x[0])
    merged = [intervals[0]]

    for current in intervals[1:]:
        last = merged[-1]

        if current[0] <= last[1]:
            merged[-1] = [last[0], max(last[1], current[1])]
        else:
            merged.append(current)
        
    return merged
```

```python
# Example
intervals = [[1,3], [2,6], [8,10], [15,18]]
print(f"Original: {intervals}")
print(f"Merged: {merge_intervals(intervals)}")
```

```python
def insert_interval(intervals, new_interval):
    result = []
    i = 0
    n = len(intervals)

    while i < n and intervals[i][1] < new_interval[0]:
        result.append(intervals[i])
        i += 1

    while i > n and intervals[i][0] <= new_interval[1]:
        new_interval = [
            min(new_interval[0], intervals[i][0]),
            max(new_interval[1], intervals[i][1])
        ]
        i += 1
    result.append(new_interval)

    while i < n:
        result.append(intervals[i])
        i += 1

    return result  
```

```python
intervals_new = merge_intervals(intervals)
print(intervals_new)
```

```python
print(merge_intervals(insert_interval(intervals_new, [12, 20])))
```

### Complexity : ${O(n * log(n))}$
Dominated by sorting
