## Pattern 18: Math Pattern

### 🧠 What are Math Patterns in DSA?

**Math Patterns** use fundamental mathematical properties and number theory to solve problems efficiently — without brute force. Instead of checking every possibility, you leverage **mathematical shortcuts** to arrive at the answer faster.

> **Think of it like knowing multiplication tables** ✖️ — You *could* add 7 seven times to get 49, but knowing `7 × 7 = 49` is instant. Math patterns give you these shortcuts for coding problems.

---

### 🔑 Two Core Concepts in This Pattern

| Concept | What It Does | Key Insight |
|---------|-------------|-------------|
| **GCD (Greatest Common Divisor)** | Finds the largest number that divides both `a` and `b` | The **Euclidean Algorithm** reduces this to simple repeated division |
| **Primality Test** | Checks if a number has only two factors: 1 and itself | You only need to check divisors up to **√n** |

---

### 🔑 The Euclidean Algorithm — Core Idea

The GCD is based on one elegant mathematical fact:

```
gcd(a, b) = gcd(b, a % b)
```

> If `a` is divisible by `b`, then `gcd(a, b) = b`
> Otherwise, the GCD of `a` and `b` is the same as the GCD of `b` and the remainder `a % b`

**Why does this work?**
If `d` divides both `a` and `b`, then `d` also divides `a - b`, and therefore `a % b`. So the GCD is preserved at each step, but the numbers get smaller — until one becomes 0.

```
Example: gcd(48, 18)

  gcd(48, 18) → 48 % 18 = 12 → gcd(18, 12)
  gcd(18, 12) → 18 % 12 = 6  → gcd(12, 6)
  gcd(12, 6)  → 12 % 6  = 0  → gcd(6, 0)
  b == 0 → return a = 6 ✅
```

---

### 🔑 The √n Primality Trick — Core Idea

Why do we only check divisors up to `√n`?

```
If n = a × b, then one of them MUST be ≤ √n.

Example: n = 36
  Factors: 1×36, 2×18, 3×12, 4×9, 6×6
                                    ↑ √36 = 6

  After 6, factors just MIRROR: 9×4, 12×3, 18×2, 36×1
  So if no factor found up to √n, none exist beyond it either!
```

```
Checking 37:
  √37 ≈ 6.08 → check 2, 3, 4, 5, 6
  37 % 2 = 1 ✗
  37 % 3 = 1 ✗
  37 % 4 = 1 ✗
  37 % 5 = 2 ✗
  37 % 6 = 1 ✗
  No divisor found → 37 is PRIME ✅
```

### 🪜 Step-by-Step Walkthrough

---

#### Function 1: `gcd(48, 18)`

```
Step 1: a=48, b=18
  b ≠ 0 → continue
  a, b = b, a%b → a, b = 18, 48%18 → a=18, b=12

Step 2: a=18, b=12
  b ≠ 0 → continue
  a, b = 12, 18%12 → a=12, b=6

Step 3: a=12, b=6
  b ≠ 0 → continue
  a, b = 6, 12%6 → a=6, b=0

Step 4: a=6, b=0
  b == 0 → STOP → return a = 6 ✅

gcd(48, 18) = 6
```

#### Another example: `gcd(56, 98)`

```
Step 1: a=56, b=98 → a, b = 98, 56%98 → a=98, b=56
  (Wait — 56 < 98, so 56%98=56. The algorithm auto-swaps!)

Step 2: a=98, b=56 → a, b = 56, 98%56 → a=56, b=42
Step 3: a=56, b=42 → a, b = 42, 56%42 → a=42, b=14
Step 4: a=42, b=14 → a, b = 14, 42%14 → a=14, b=0
Step 5: b == 0 → return 14 ✅

gcd(56, 98) = 14
```

---

#### Function 2: `is_prime(29)`

```
Step 1: n=29, n < 2? No → continue
Step 2: Check range(2, int(29**0.5) + 1) = range(2, 6) → check 2, 3, 4, 5

  i=2: 29 % 2 = 1 → not divisible
  i=3: 29 % 3 = 2 → not divisible
  i=4: 29 % 4 = 1 → not divisible
  i=5: 29 % 5 = 4 → not divisible

Step 3: No divisor found → return True ✅
29 is PRIME!
```

#### Non-prime example: `is_prime(15)`

```
Step 1: n=15, n < 2? No → continue
Step 2: range(2, int(15**0.5) + 1) = range(2, 4) → check 2, 3

  i=2: 15 % 2 = 1 → not divisible
  i=3: 15 % 3 = 0 → DIVISIBLE! → return False ❌

15 is NOT prime (15 = 3 × 5)
```

#### Edge cases: `is_prime(0)`, `is_prime(1)`, `is_prime(2)`

```
is_prime(0): 0 < 2 → return False ❌
is_prime(1): 1 < 2 → return False ❌
is_prime(2): 2 < 2? No → range(2, 2) is empty → return True ✅
             (2 is the smallest and only even prime!)
```

### 🐍 Code Walkthrough

#### `gcd(a, b)` — Euclidean Algorithm
- Repeatedly replaces `(a, b)` with `(b, a % b)`
- When `b` becomes `0`, `a` holds the GCD
- The `while b:` loop is elegant — `0` is falsy in Python, so it stops automatically
- Works correctly even if `a < b` (auto-swaps on the first iteration)

#### `is_prime(n)` — Trial Division
- Returns `False` for `n < 2` (0 and 1 are not prime)
- Checks all numbers from `2` to `√n` (inclusive)
- If any divides `n` evenly → not prime, return `False`
- If the loop completes without finding a divisor → prime, return `True`

````python
def gcd(a, b):
    while b:
        a, b = b, a % b
    return a
````

````python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
````

### ▶️ Example Usage

````python
# === GCD Examples ===
print("=== GCD (Greatest Common Divisor) ===\n")
gcd_tests = [(48, 18), (56, 98), (100, 75), (17, 13), (0, 5), (12, 12)]
for a, b in gcd_tests:
    result = gcd(a, b)
    print(f"  gcd({a}, {b}) = {result}")

# LCM using GCD:  lcm(a,b) = (a * b) // gcd(a,b)
print("\n=== LCM using GCD ===\n")
def lcm(a, b):
    return (a * b) // gcd(a, b)

lcm_tests = [(4, 6), (12, 18), (7, 5), (15, 20)]
for a, b in lcm_tests:
    result = lcm(a, b)
    print(f"  lcm({a}, {b}) = {result}")

# === Prime Examples ===
print("\n=== Primality Test ===\n")
prime_tests = [0, 1, 2, 3, 4, 7, 10, 13, 17, 25, 29, 97, 100]
for n in prime_tests:
    result = is_prime(n)
    symbol = "✅ Prime" if result else "❌ Not Prime"
    print(f"  is_prime({n:>3}) → {symbol}")

# Find all primes up to 50
print("\n=== All Primes up to 50 ===\n")
primes = [n for n in range(2, 51) if is_prime(n)]
print(f"  {primes}")
print(f"  Count: {len(primes)} primes")

# === Bonus: GCD of a list of numbers ===
print("\n=== GCD of Multiple Numbers ===\n")
def gcd_list(nums):
    result = nums[0]
    for num in nums[1:]:
        result = gcd(result, num)
    return result

print(f"  gcd([12, 18, 24]) = {gcd_list([12, 18, 24])}")
print(f"  gcd([100, 75, 50, 25]) = {gcd_list([100, 75, 50, 25])}")
````

### ⏱️ Time & Space Complexity

#### `gcd(a, b)` — Euclidean Algorithm

| Aspect | Complexity | Explanation |
|--------|-----------|-------------|
| **Time** | **`O(log(min(a, b)))`** | Each step reduces the larger number by at least half. The number of steps is proportional to the number of digits |
| **Space** | **`O(1)`** | Only uses two variables `a` and `b`, modified in-place |

> 💡 This is **incredibly fast**. `gcd(10^18, 10^18)` takes at most ~60 steps!

#### `is_prime(n)` — Trial Division

| Aspect | Complexity | Explanation |
|--------|-----------|-------------|
| **Time** | **`O(√n)`** | Checks divisors from 2 to √n. For `n = 1,000,000`, that's only ~1000 checks |
| **Space** | **`O(1)`** | Only uses loop variable `i` |

---

### 📊 Summary Table

| Function | Time | Space | Core Trick |
|----------|------|-------|------------|
| `gcd(a, b)` | `O(log(min(a,b)))` | `O(1)` | `gcd(a,b) = gcd(b, a%b)` — remainder shrinks fast |
| `is_prime(n)` | `O(√n)` | `O(1)` | If `n = a×b`, one factor must be `≤ √n` |

### 🌍 Classic Math-Based DSA Problems

| Problem | Technique |
|---------|-----------|
| **GCD of Array** | Apply `gcd` pairwise across all elements |
| **LCM of Array** | `lcm(a,b) = a*b // gcd(a,b)` pairwise |
| **Count Primes (Sieve)** | Sieve of Eratosthenes — `O(n log log n)` |
| **Power (Fast Exponentiation)** | `a^n` in `O(log n)` using squaring |
| **Missing Number** | Sum formula or XOR trick |
| **Ugly Numbers** | Numbers whose prime factors are only 2, 3, 5 |
| **Fraction to Decimal** | Long division + hash map for cycle detection |
| **Happy Number** | Detect cycle in digit-square sequence |

### 💡 Key Takeaway

> **Math Patterns = "Use mathematical properties to avoid brute force."**
> The two pillars are: **Euclidean Algorithm** (`gcd(a,b) = gcd(b, a%b)`) for divisor problems, and the **√n trick** for primality — both turn potentially expensive computations into lightning-fast solutions.
