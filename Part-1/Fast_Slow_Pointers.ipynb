# Pattern 3: Fast & Slow Pointers (Floyd's Cycle Detection)

**Problem:** Detect if a linked list has a cycle.

**Solution Approach:**
Use two pointers moving at different speeds. The slow pointer moves one step at a time, while the fast pointer moves two steps. If there's a cycle, they will eventually meet. Also known as 'Tortoise and Hare' algorithm.

## Python Code:

```python
class ListNode:
    '''Definition of singly linked list node'''
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def has_cycle(head):
    """
    Detect cycle in linked list using Floyd's algorithm.
    Args:
    head: Head node of linked list
    Returns:
    True if cycle exists, False otherwise
    Example:
    1 -> 2 -> 3 -> 4 -> 2 (cycle back to node 2)
    Returns: True
    """
    if not head or not head.next:
        return False
    
    slow = head
    fast = head

    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next

        if slow == fast:
            return True
    
    return False
```

```python
def find_cycle(head):
    """
    Find the node where cycle begins.
    After detecting cycle, reset one pointer to head.
    Move both pointers one step at a time.
    They will meet at cycle start.
    """
    if not(head):
        return None
    
    slow = fast = head
    has_cycle_flag = False

    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        if slow == fast:
            has_cycle_flag = True
            break
    
    if not has_cycle_flag:
        return None
    
    # Find cycle start
    slow = head
    while slow != fast:
        slow = slow.next
        fast = fast.next
    
    return slow
```

```python
def find_middle_of_linked_list(head):
    """
    Find middle node of linked list.
    Another application of fast & slow pointers.
    When fast reaches end, slow is at middle.
    """
    if not head:
        return None
    
    slow = fast = head

    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next

    return slow
```

```python
# Example usage
# Creating a linked list with cycle: 1 -> 2 -> 3 -> 4 -> 2
node1 = ListNode(1)
node2 = ListNode(2)
node3 = ListNode(3)
node4 = ListNode(4)
node1.next = node2
node2.next = node3
node3.next = node4
node4.next = node2 # Creates cycle
print("Cycle Detection:")
print(f"Has cycle: {has_cycle(node1)}") # True
# Creating list without cycle: 1 -> 2 -> 3 -> 4
node1_no_cycle = ListNode(1)
node2_no_cycle = ListNode(2)
node3_no_cycle = ListNode(3)
node4_no_cycle = ListNode(4)
node1_no_cycle.next = node2_no_cycle
node2_no_cycle.next = node3_no_cycle
node3_no_cycle.next = node4_no_cycle
print(f"Has cycle (no cycle list): {has_cycle(node1_no_cycle)}") # False
```

**Complexity Analysis:**
* Time Complexity: ${O(n)}$ - At most ${2n}$ steps (${n}$ to detect, ${n}$ to find start)
* Space Complexity: ${O(1)}$ - Only two pointers used

**When to Use:**
* Cycle detection in linked lists
* Finding middle of linked list
* Detecting duplicate numbers in array (mathematical variation)

**Key Applications:**
* Linked List Cycle (detect, find start)
* Happy Number problem
* Find middle element
* Palindrome Linked List
